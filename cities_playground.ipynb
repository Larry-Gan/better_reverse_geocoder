```python
# --- Configuration ---

# Output file
GEOJSON_OUTPUT_PATH = 'cities_merged.geojson'

# Parameters
RADIUS_METERS = 15000  # 15 km

def merge_city_data(gdf_cities, df_worldcities, output_path, radius):
    """
    Adds cities from a CSV to a GeoJSON, cleaning up redundant entries within a radius.

    Args:
        gdf_cities (str): geojson to work on.
        df_worldcities (str): csv to df.
        output_path (str): Path to save the merged GeoJSON file.
        radius (int): The radius in meters to search for redundant cities.
    """
    print("Step 2: Preparing GeoDataFrames...")
    # Create a GeoDataFrame from the world cities CSV
    geometry = [Point(xy) for xy in zip(df_worldcities.lng, df_worldcities.lat)]
    gdf_worldcities = gpd.GeoDataFrame(df_worldcities, geometry=geometry, crs="EPSG:4326")

    # To calculate distances in meters, we need to project our data to a suitable CRS.
    # EPSG:3857 (Web Mercator) is a common choice for this.
    gdf_cities_proj = gdf_cities.to_crs(epsg=3857)
    gdf_worldcities_proj = gdf_worldcities.to_crs(epsg=3857)

    # We'll keep track of original indices to drop and new rows to add
    indices_to_drop = set()
    new_city_rows = []

    print(f"Step 3: Processing {len(gdf_worldcities)} new cities...")
    # Iterate through each new city from the CSV to add it
    for idx, new_city in gdf_worldcities.iterrows():
        new_city_proj = gdf_worldcities_proj.loc[idx]

        # --- a) Find redundant cities to remove ---
        # Create a buffer (circle) around the new city
        buffer = new_city_proj.geometry.buffer(radius)

        # Find original cities that are within the buffer AND have the same country code
        mask_within_buffer = gdf_cities_proj.geometry.within(buffer)
        mask_same_country = gdf_cities_proj['city_country_code'] == new_city['iso2']
        
        redundant_indices = gdf_cities_proj[mask_within_buffer & mask_same_country].index
        indices_to_drop.update(redundant_indices)

        # --- b) Find nearest original city to inherit admin names ---
        # Use original CRS data for this, as we just need the closest point, not a specific distance
        all_original_points = gdf_cities.geometry.unary_union
        if all_original_points.is_empty:
            print(f"Warning: Original GeoDataFrame is empty. Cannot find a nearest neighbor for {new_city['city']}.")
            nearest_original_city = pd.Series({'city_admin1_name': None, 'city_admin2_name': None})
        else:
            _, nearest_geom = nearest_points(new_city.geometry, all_original_points)
            # Find the original city that corresponds to this nearest geometry
            nearest_original_city = gdf_cities[gdf_cities.geometry == nearest_geom].iloc[0]

        # --- c) Prepare the new row with the correct schema ---
        new_row = {
            'city_name': new_city['city_ascii'],
            'city_latitude': new_city['lat'],
            'city_longitude': new_city['lng'],
            'city_country_code': new_city['iso2'],
            'city_admin1_name': nearest_original_city['city_admin1_name'],
            'city_admin2_name': nearest_original_city['city_admin2_name'],
            'geometry': new_city.geometry
        }
        new_city_rows.append(new_row)

        if idx % 100 == 0:
            print(f"  Processed {idx + 1}/{len(gdf_worldcities)} new cities...")
            
    print("\nStep 4: Merging the data...")
    # Drop the redundant rows from the original dataframe
    print(f"  Found {len(indices_to_drop)} redundant cities to remove.")
    gdf_cities_pruned = gdf_cities.drop(index=list(indices_to_drop))
    print(f"  Original cities remaining: {len(gdf_cities_pruned)}")

    # Create a new GeoDataFrame for the cities we're adding
    gdf_new_cities = gpd.GeoDataFrame(new_city_rows, crs="EPSG:4326")
    print(f"  New cities to add: {len(gdf_new_cities)}")

    # Combine the pruned original data with the new data
    gdf_merged = pd.concat([gdf_cities_pruned, gdf_new_cities], ignore_index=True)

    print(f"\nStep 5: Saving merged file to {output_path}...")
    # Save the final result
    gdf_merged.to_file(output_path, driver='GeoJSON')
    print("--- Process Complete! ---")
    print(f"Final dataset contains {len(gdf_merged)} cities.")

    return gdf_merged


# The sample "geojson" in the prompt looks like a tab-separated file.
# The code includes a fallback to handle this format if geopandas fails to read it as a standard GeoJSON.
merge_city_data(
    gdf_cities=df,
    df_worldcities=df2,
    output_path=GEOJSON_OUTPUT_PATH,
    radius=RADIUS_METERS
)
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# This new helper function can be placed before your main function
def normalize_text(text):
    """
    Converts a string to its closest ASCII representation.
    Returns the original if input is not a string (e.g., None or NaN).
    """
    if isinstance(text, str):
        return unidecode(text)
    return text

def merge_and_update_cities(gdf_cities, df_worldcities, radius_meters=15000):
    """
    Merges and updates city data based on proximity.

    Instead of deleting, this function updates existing entries or adds new ones.
    - If an existing city (from gdf_cities) is within the radius of a new city
      (from df_worldcities), it gets a 'nearby_large_city' tag.
    - If a new city has no existing cities within its radius, it is added as a new row.

    Args:
        gdf_cities (gpd.GeoDataFrame): The original GeoDataFrame of cities.
        df_worldcities (pd.DataFrame): A DataFrame with the new, "cleaner" city data.
        radius_meters (int): The radius in meters to search for nearby cities.

    Returns:
        tuple: A tuple containing three GeoDataFrames:
               1. gdf_final_merged: The complete, merged, and updated dataset.
               2. gdf_updated_rows: A GDF of only the original rows that were updated.
               3. gdf_added_rows: A GDF of only the new rows that were added.
    """
    print("Step 1: Preparing data and initializing columns...")
    # Work on a copy to avoid modifying the original DataFrame passed to the function
    gdf_cities_updated = gdf_cities.copy()
    df_worldcities_cleaned = df_worldcities.copy()

    # --- ADD THIS BLOCK ---
    # Normalize text columns for consistent matching and clean output
    print("  Normalizing names to ASCII...")
    gdf_cities_updated['city_name'] = gdf_cities_updated['city_name'].apply(normalize_text)
    gdf_cities_updated['city_admin1_name'] = gdf_cities_updated['city_admin1_name'].apply(normalize_text)
    
    df_worldcities_cleaned['city_ascii'] = df_worldcities_cleaned['city_ascii'].apply(normalize_text)
    df_worldcities_cleaned['admin_name'] = df_worldcities_cleaned['admin_name'].apply(normalize_text)
    # --- END OF ADDED BLOCK ---
    
    # Add the new column and initialize it with a null value (pandas' NA is preferred)
    gdf_cities_updated['nearby_large_city'] = pd.NA

    # Create a GeoDataFrame from the world cities DataFrame
    geometry = [Point(xy) for xy in zip(df_worldcities_cleaned.lng, df_worldcities_cleaned.lat)]
    gdf_worldcities = gpd.GeoDataFrame(df_worldcities_cleaned, geometry=geometry, crs="EPSG:4326")

    print("Step 2: Projecting data for accurate distance calculation...")
    # Use a projected CRS (like Web Mercator) to work with meters
    gdf_cities_proj = gdf_cities_updated.to_crs(epsg=3857)
    gdf_worldcities_proj = gdf_worldcities.to_crs(epsg=3857)

    # Keep track of which original rows were updated and which new rows to add
    updated_indices = set()
    new_city_rows = []

    print(f"Step 3: Processing {len(gdf_worldcities)} new cities...")
    # Iterate through each "clean" city from the CSV
    for idx, new_city in gdf_worldcities.iterrows():
        new_city_proj = gdf_worldcities_proj.loc[idx]

        # Create a buffer (circle) around the new city to define the search area
        buffer = new_city_proj.geometry.buffer(radius_meters)

        # Find original cities within the buffer and with the same country code
        mask_within_buffer = gdf_cities_proj.geometry.within(buffer)
        mask_same_country = gdf_cities_proj['city_country_code'] == new_city['iso2']
        
        nearby_original_indices = gdf_cities_proj[mask_within_buffer & mask_same_country].index

        if not nearby_original_indices.empty:
            # --- LOGIC A: FOUND NEARBY CITY ---
            # Update the 'nearby_large_city' column for the found cities
            gdf_cities_updated.loc[nearby_original_indices, 'nearby_large_city'] = new_city['city_ascii']
            # Record the indices of the rows we modified
            updated_indices.update(nearby_original_indices)
        else:
            # --- LOGIC B: NO NEARBY CITY FOUND ---
            # Add the new city from the CSV as a new row to our dataset
            new_row = {
                'city_name': new_city['city_ascii'],
                'city_latitude': new_city['lat'],
                'city_longitude': new_city['lng'],
                'city_country_code': new_city['iso2'],
                'city_admin1_name': new_city['admin_name'], # Map admin_name to city_admin1_name
                'city_admin2_name': None, # No source for admin2, so set to None
                'geometry': new_city.geometry,
                'nearby_large_city': pd.NA # This city is the 'large city', not near one
            }
            new_city_rows.append(new_row)

        if (idx + 1) % 100 == 0:
            print(f"  Processed {idx + 1}/{len(gdf_worldcities)} new cities...")

    print("\nStep 4: Assembling final GeoDataFrames...")
    # Create a GeoDataFrame for the newly added cities
    gdf_added_rows = gpd.GeoDataFrame(new_city_rows, crs="EPSG:4326")
    print(f"  Found {len(updated_indices)} original cities to update.")
    print(f"  Found {len(gdf_added_rows)} new cities to add.")

    # Get a dataframe of only the rows that were modified
    gdf_updated_rows = gdf_cities_updated.loc[list(updated_indices)].copy()

    # Combine the (now updated) original data with the new city rows
    gdf_final_merged = pd.concat([gdf_cities_updated, gdf_added_rows], ignore_index=True)

    print("\n--- Process Complete! ---")
    print(f"Final dataset contains {len(gdf_final_merged)} cities.")
    
    return gdf_final_merged, gdf_updated_rows, gdf_added_rows



# --- Call the Function ---
final_gdf, updated_gdf, added_gdf = merge_and_update_cities(
    gdf_cities=df,
    df_worldcities=df2,
    radius_meters=15000
)

# --- Handle and Inspect the Outputs ---
print("\n\n--- OUTPUTS ---")

print("\n1. Final Merged GeoDataFrame (first 5 rows):")
print(final_gdf.head())

print("\n2. Rows from Original GDF that were UPDATED:")
print(updated_gdf)

print("\n3. Rows that were NEWLY ADDED:")
print(added_gdf)

# --- Save the results to files ---
print("\nSaving files...")
final_gdf.to_file("cities_final_merged.geojson", driver='GeoJSON')
updated_gdf.to_csv("cities_updated_rows.csv", index=False)
added_gdf.to_csv("cities_added_rows.csv", index=False)
print("Files saved successfully.")
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from unidecode import unidecode

def normalize_text(text):
    """
    Converts a string to its closest ASCII representation.
    """
    if isinstance(text, str):
        return unidecode(text)
    return text

def merge_and_update_cities(gdf_cities, df_worldcities, radius_meters=15000):
    """
    Merges and updates city data based on proximity, ensuring the closest city is chosen.

    - Uses sjoin_nearest to guarantee the closest "large city" is assigned.
    - If an existing city is within the radius of a new city, it gets a 'nearby_large_city' tag.
    - If a new city has no existing cities within its radius, it is added as a new row.
    """
    print("Step 1: Preparing data and normalizing names...")
    gdf_cities_normalized = gdf_cities.copy()
    df_worldcities_cleaned = df_worldcities.copy()

    # Normalize text columns for consistent matching
    gdf_cities_normalized['city_name'] = gdf_cities_normalized['city_name'].apply(normalize_text)
    gdf_cities_normalized['city_admin1_name'] = gdf_cities_normalized['city_admin1_name'].apply(normalize_text)
    
    df_worldcities_cleaned['city_ascii'] = df_worldcities_cleaned['city_ascii'].apply(normalize_text)
    df_worldcities_cleaned['admin_name'] = df_worldcities_cleaned['admin_name'].apply(normalize_text)

    # Create a GeoDataFrame from the world cities DataFrame
    geometry = [Point(xy) for xy in zip(df_worldcities_cleaned.lng, df_worldcities_cleaned.lat)]
    gdf_worldcities = gpd.GeoDataFrame(df_worldcities_cleaned, geometry=geometry, crs="EPSG:4326")

    print("Step 2: Projecting data for accurate distance calculation...")
    gdf_cities_proj = gdf_cities_normalized.to_crs(epsg=3857)
    gdf_worldcities_proj = gdf_worldcities.to_crs(epsg=3857)

    # --- REVISED LOGIC: Use sjoin_nearest to find the closest match ---
    print("Step 3: Finding closest large city for each original city...")
    
    # This single operation replaces the entire for-loop.
    # It finds the nearest city in gdf_worldcities_proj for each city in gdf_cities_proj,
    # but only if it's within radius_meters.
    # 'how=left' ensures we keep all original cities.
    gdf_cities_updated = gpd.sjoin_nearest(
        gdf_cities_proj, 
        gdf_worldcities_proj, 
        max_distance=radius_meters, 
        how='left',
        rsuffix='large' # Suffix for columns from the right GeoDataFrame
    )
    
    # Rename the joined column to 'nearby_large_city'
    # The name is 'city_ascii_large' because of the 'rsuffix'
    gdf_cities_updated.rename(columns={'city_ascii': 'nearby_large_city'}, inplace=True)
    
    # Select only the columns we want to keep, cleaning up extra columns from the join
    # (like index_large, lat_large, country_large, etc.)
    final_cols = list(gdf_cities_normalized.columns) + ['nearby_large_city']
    gdf_cities_updated = gdf_cities_updated[final_cols]

    # Convert back to the original CRS
    gdf_cities_updated = gdf_cities_updated.to_crs("EPSG:4326")


    print("\nStep 4: Identifying new cities to add...")
    # Find which large cities were successfully matched to an original city
    matched_large_city_indices = gdf_cities_updated[gdf_cities_updated['nearby_large_city'].notna()].index

    # To find which worldcities were used, we need to go back to the join result
    join_result = gpd.sjoin_nearest(gdf_cities_proj, gdf_worldcities_proj, max_distance=radius_meters, how='inner')
    used_worldcities_indices = join_result['index_right'].unique()

    # The cities to add are those in gdf_worldcities whose indices were NOT used in the join
    new_cities_to_add_df = gdf_worldcities[~gdf_worldcities.index.isin(used_worldcities_indices)]
    print(f"  Found {len(new_cities_to_add_df)} new cities to add.")

    new_city_rows = []
    for _, new_city in new_cities_to_add_df.iterrows():
        new_row = {
            'city_name': new_city['city_ascii'],
            'city_latitude': new_city['lat'],
            'city_longitude': new_city['lng'],
            'city_country_code': new_city['iso2'],
            'city_admin1_name': new_city['admin_name'],
            'city_admin2_name': None,
            'geometry': new_city.geometry,
            'nearby_large_city': pd.NA 
        }
        new_city_rows.append(new_row)

    print("\nStep 5: Assembling final GeoDataFrames...")
    gdf_added_rows = gpd.GeoDataFrame(new_city_rows, crs="EPSG:4326")

    # Get a dataframe of only the rows that were modified
    gdf_updated_rows = gdf_cities_updated[gdf_cities_updated['nearby_large_city'].notna()].copy()
    print(f"  Found {len(gdf_updated_rows)} original cities to update.")

    # Combine the (now updated) original data with the new city rows
    gdf_final_merged = pd.concat([gdf_cities_updated, gdf_added_rows], ignore_index=True)

    print("\n--- Process Complete! ---")
    print(f"Final dataset contains {len(gdf_final_merged)} cities.")
    
    return gdf_final_merged, gdf_updated_rows, gdf_added_rows

# --- Example Usage (assuming df and df2 are pre-loaded) ---
final_gdf, updated_gdf, added_gdf = merge_and_update_cities(
    gdf_cities=df,
    df_worldcities=df2,
    radius_meters=15000
)
```
